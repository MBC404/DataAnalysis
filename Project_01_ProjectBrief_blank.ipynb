# The Best Movies And The Worst Movies Part 1

```python
import pandas as pd
from IPython.display import HTML
pd.options.display.max_columns = 30
pd.options.display.max_colwidth = 200
movies = pd.read_csv("movies_complete.csv",parse_dates=["release_date"])
```

## Adding Profit column and ROI column to analyze 

```python
movies["Profit_musd"] = movies.revenue_musd.sub(movies.budget_musd)
movies["ROI"] = movies.revenue_musd.div(movies.budget_musd)
movies
```

### Making separate dataframe for best movies

```python
movies_best = movies[["poster_path","title","budget_musd","revenue_musd","vote_count","vote_average","popularity"]].copy()
movies_best["Profit"] = movies_best.revenue_musd.sub(movies_best.budget_musd)
movies_best["ROI"] = movies_best.revenue_musd.div(movies_best.budget_musd)
```

```python
movies_best.head(11)
```

```python
movies_best.columns = ["","TITLE","Budget","Revenue","Votes","VoteAverage","popularity","Profit","ROI"]
```

```python
movies_best.set_index("TITLE",inplace = True)
```

```python
movies_best.iloc[0,0]
```

```python
subset = movies_best.iloc[:5,:2]
HTML(subset.to_html(escape = False))
```

```python
HTML(movies_best.loc[movies_best.Budget >= 5].sort_values(by = "ROI" , ascending = False).head(5).to_html(escape = False))
```

```python
HTML(movies_best.loc[movies_best.Votes >= 126].sort_values(by = "VoteAverage",ascending = False).head(5).to_html(escape = False))
```

```python
movies_best["Budget"] = movies_best["Budget"].fillna(0)
movies_best["Votes"] = movies_best["Votes"].fillna(0)
```

```python
def best_worst(n,by,ascending = False,minbud = 0,minvotes = 0):
    df2 = movies_best.loc[(movies_best.Budget >= minbud) & (movies_best.Votes >= minvotes),
                             ["",by]].sort_values(by = by,ascending = ascending).head(n).copy()
    return HTML(df2.to_html(escape = False))
```

```python
best_worst(n = 5,by = "Revenue")
```

```python
best_worst(n = 5,by = "Profit")
```

```python
best_worst(n = 5,by = "Profit",ascending = True)
```

```python
best_worst(n = 5,by = "ROI",ascending = False,minbud = 40,minvotes = 10)
```

```python
best_worst(n = 5,by = "ROI",ascending = True,minbud = 40,minvotes = 10)
```

```python
best_worst(5,"Budget")
```

```python
gen = gen.groupby("GENRE").agg(TotalMovies = ("title","count"),
                          Total_rev = ("revenue_musd","sum"),
                           Mean_rev = ("revenue_musd","mean"),
                            Popularity = ("popularity","mean"),
                             )
```
